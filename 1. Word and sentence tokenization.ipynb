```python
text =  ''' Cats are predators. Cats do not purr while communicating with each other. 
Cats purr only at their young ones. They have developed verbal purrs to communicate with humans'''
```

```python
# Python inbuilt method

word = text.split()
sent = text.split(". ")

print(" Word tokens : ", word)
print(" Sentence tokens : ", sent)
```

```python
# RegEX

import re

word = re.findall("[\w]+", text)
sent = re.compile("[.!?] ").split(text)

print(" Word tokens : ", word)
print(" Sentence tokens : ", sent)
```

```python
# NLTK

import nltk
from nltk.tokenize import word_tokenize, sent_tokenize

word = word_tokenize(text)
sent = sent_tokenize(text)

print(" Word tokens : ", word)
print(" Sentence tokens : ", sent)
```

```python
# Spacy

import spacy
nlp = spacy.load("en_core_web_sm")
doc = nlp(text)
word = []
for token in doc :
    word.append(token.text)

sent = []
for token in doc.sents :
    sent.append(token.text)

print(" Word tokens : ", word)
print(" Sentence tokens : ", sent)   
```

```python
# Keras

from keras.preprocessing.text import text_to_word_sequence

word = text_to_word_sequence(text)
sent = text_to_word_sequence(text, split = ".", filters = "!?\n")

print(" Word tokens : ", word)
print(" Sentence tokens : ", sent)
```
